简易感知器

```python
import numpy as np
```

```python
class Linear:
    def __init__(self, dim):
        self.weight = [0.0 for _ in range(dim)]
        self.bias = 0.0
    def __str__(self):
        return "weight={}\nbias={}\n".format(self.weight, self.bias)
    def __forward(self, sample):
        return np.dot(sample, self.weight) + self.bias
    def __activation(self, x):
        #sigmod
        return x
    def __update(self, sample, output, label, learnig_rate):
        delta = label - output
        self.weight = np.add(self.weight, learnig_rate * delta * sample)
        self.bias += learnig_rate * delta
        print(self.__str__())
    def predict(self, sample):
        return self.__activation(self.__forward(sample))
    def train(self, input, labels, epochs, learnig_rate):
        for _ in range(epochs):
            for sample, label in zip(input, labels):
                output = self.predict(sample)
                self.__update(sample, output, label, learnig_rate)
```

```python
input = np.array([[5], [3], [8], [1.4], [10.1]])
label = np.array([5500, 2300, 7600, 1800, 11400])

model = Linear(1)
model.train(input, label, 10, 0.01)
```

```python
print('Work 3.4 years, monthly salary = %.2f' % model.predict([3.4]))
print('Work 15 years, monthly salary = %.2f' % model.predict([15]))
print('Work 1.3 years, monthly salary = %.2f' % model.predict([1.5]))
print('Work 6.5 years, monthly salary = %.2f' % model.predict([6.5]))
```

```python
import matplotlib.pyplot as plt
plt.scatter(input, label, c="red")
plt.plot(input, model.weight * input + model.bias, c="green")
plt.xlabel("work years")
plt.ylabel("salary")
plt.title("salary of work years")
plt.show()
```
